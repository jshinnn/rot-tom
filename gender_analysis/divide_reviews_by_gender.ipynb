```python
# Then we should somehow analyze the gender and the genre and the sentiment of the review
import pandas as pd

# Import the CSV file into a pandas DataFrame
df = pd.read_csv("reviews-1.csv")

# Create two new DataFrames - one for the reviews done by females or mostly females and one for the reviews done by males or mostly males
female_reviews_df = pd.DataFrame(columns=df.columns)
male_reviews_df = pd.DataFrame(columns=df.columns)

# Filter the original DataFrame by the "gender" column to get the rows where the gender is female or mostly female
female_reviews = df[(df["gender"] == "female") | (df["gender"] == "mostly_female")]
female_reviews_df = pd.concat([female_reviews_df, female_reviews])

# Filter the original DataFrame by the "gender" column to get the rows where the gender is male or mostly male
male_reviews = df[(df["gender"] == "male") | (df["gender"] == "mostly_male")]
male_reviews_df = pd.concat([male_reviews_df, male_reviews])

female_reviews.to_csv('female_reviews.csv', index=False)
male_reviews.to_csv('male_reviews.csv', index=False)
```
