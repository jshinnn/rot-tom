```python
# Extracting the link, title, genre list, tomatometer / audience status, and dates of release from movie metadata
import pandas as pd

df = pd.read_csv('../raw_data/og_metadata_with_genre_list.csv')
df['year_of_release'] = pd.to_datetime(df['original_release_date'], format='%Y-%m-%d')
df['year_of_release'] = df['year_of_release'].dt.strftime('%Y')
# Fill in missing values with zero
df['year_of_release'] = df['year_of_release'].fillna(0)
df['year_of_release'] = df['year_of_release'].astype(int)

df['year_of_streaming_release'] = pd.to_datetime(df['streaming_release_date'], format='%Y-%m-%d')
df['year_of_streaming_release'] = df['year_of_streaming_release'].dt.strftime('%Y')
# Fill in missing values with zero
df['year_of_streaming_release'] = df['year_of_streaming_release'].fillna(0)
df['year_of_streaming_release'] = df['year_of_streaming_release'].astype(int)

df = df[['rotten_tomatoes_link', 
        'movie_title', 
        'year_of_release',
        'year_of_streaming_release',
        'tomatometer_status',
        'audience_status',
        'genre_list']]

df.to_csv('metadata_temp.csv')
```

```python
import csv

# this creates a csv without the first column and its values
with open('metadata_temp.csv', 'r') as input_file, open('metadata_new.csv', 'w', newline='') as output_file:
    reader = csv.reader(input_file)
    writer = csv.writer(output_file)
    for row in reader:
        writer.writerow(row[1:])
```

```python
# Extracting link, critic name, review type, review year, gender from critic reviews
import pandas as pd

df_1 = pd.read_csv('../clean_data/critic_reviews-1.csv')
df_1['review_year'] = pd.to_datetime(df_1['review_date'], format='%Y-%m-%d')
df_1['review_year'] = df_1['review_year'].dt.strftime('%Y')
# Fill in missing values with zero
df_1['review_year'] = df_1['review_year'].fillna(0)
df_1['review_year'] = df_1['review_year'].astype(int)
df_1 = df_1[['rotten_tomatoes_link', 
            'critic_name',
            'review_type',
            'review_year',
            'gender']]

df_2 = pd.read_csv('../clean_data/critic_reviews-2.csv')
df_2['review_year'] = pd.to_datetime(df_2['review_date'], format='%Y-%m-%d')
df_2['review_year'] = df_2['review_year'].dt.strftime('%Y')
# Fill in missing values with zero
df_2['review_year'] = df_2['review_year'].fillna(0)
df_2['review_year'] = df_2['review_year'].astype(int)
df_2 = df_2[['rotten_tomatoes_link', 
            'critic_name',
            'review_type',
            'review_year',
            'gender']]

df_3 = pd.read_csv('../clean_data/critic_reviews-3.csv')
df_3['review_year'] = pd.to_datetime(df_3['review_date'], format='%Y-%m-%d')
df_3['review_year'] = df_3['review_year'].dt.strftime('%Y')
# Fill in missing values with zero
df_3['review_year'] = df_3['review_year'].fillna(0)
df_3['review_year'] = df_3['review_year'].astype(int)
df_3 = df_3[['rotten_tomatoes_link', 
            'critic_name',
            'review_type',
            'review_year',
            'gender']]

df_4 = pd.read_csv('../clean_data/critic_reviews-4.csv')
df_4['review_year'] = pd.to_datetime(df_4['review_date'], format='%Y-%m-%d')
df_4['review_year'] = df_4['review_year'].dt.strftime('%Y')
# Fill in missing values with zero
df_4['review_year'] = df_4['review_year'].fillna(0)
df_4['review_year'] = df_4['review_year'].astype(int)
df_4 = df_4[['rotten_tomatoes_link', 
            'critic_name',
            'review_type',
            'review_year',
            'gender']]

df_5 = pd.read_csv('../clean_data/critic_reviews-5.csv')
df_5['review_year'] = pd.to_datetime(df_5['review_date'], format='%Y-%m-%d')
df_5['review_year'] = df_5['review_year'].dt.strftime('%Y')
# Fill in missing values with zero
df_5['review_year'] = df_5['review_year'].fillna(0)
df_5['review_year'] = df_5['review_year'].astype(int)
df_5 = df_5[['rotten_tomatoes_link', 
            'critic_name',
            'review_type',
            'review_year',
            'gender']]

df_1.to_csv('reviews-1.csv')
df_2.to_csv('reviews-2.csv')
df_3.to_csv('reviews-3.csv')
df_4.to_csv('reviews-4.csv')
df_5.to_csv('reviews-5.csv')
```

```python
# Now merge the critic reviews with the movie metadata
import pandas as pd 

reviews_1 = pd.read_csv('reviews-1.csv')
reviews_2 = pd.read_csv('reviews-2.csv')
reviews_3 = pd.read_csv('reviews-3.csv')
reviews_4 = pd.read_csv('reviews-4.csv')
reviews_5 = pd.read_csv('reviews-5.csv')
metadata = pd.read_csv('metadata_new.csv')

merged_1 = pd.merge(reviews_1, metadata, on='rotten_tomatoes_link')
merged_2 = pd.merge(reviews_2, metadata, on='rotten_tomatoes_link')
merged_3 = pd.merge(reviews_3, metadata, on='rotten_tomatoes_link')
merged_4 = pd.merge(reviews_4, metadata, on='rotten_tomatoes_link')
merged_5 = pd.merge(reviews_5, metadata, on='rotten_tomatoes_link')

merged_1 = merged_1[['rotten_tomatoes_link', 
            'movie_title', 
            'year_of_release',
            'year_of_streaming_release',
            'tomatometer_status',
            'audience_status',
            'genre_list',
            'critic_name',
            'review_type',
            'review_year',
            'gender']]
merged_2 = merged_2[['rotten_tomatoes_link', 
            'movie_title', 
            'year_of_release',
            'year_of_streaming_release',
            'tomatometer_status',
            'audience_status',
            'genre_list',
            'critic_name',
            'review_type',
            'review_year',
            'gender']]
merged_3 = merged_3[['rotten_tomatoes_link', 
            'movie_title', 
            'year_of_release',
            'year_of_streaming_release',
            'tomatometer_status',
            'audience_status',
            'genre_list',
            'critic_name',
            'review_type',
            'review_year',
            'gender']]
merged_4 = merged_4[['rotten_tomatoes_link', 
            'movie_title', 
            'year_of_release',
            'year_of_streaming_release',
            'tomatometer_status',
            'audience_status',
            'genre_list',
            'critic_name',
            'review_type',
            'review_year',
            'gender']]
merged_5 = merged_5[['rotten_tomatoes_link', 
            'movie_title', 
            'year_of_release',
            'year_of_streaming_release',
            'tomatometer_status',
            'audience_status',
            'genre_list',
            'critic_name',
            'review_type',
            'review_year',
            'gender']]
                       
merged_1.to_csv('metadata_and_reviews-1.csv')
merged_2.to_csv('metadata_and_reviews-2.csv')
merged_3.to_csv('metadata_and_reviews-3.csv')
merged_4.to_csv('metadata_and_reviews-4.csv')
merged_5.to_csv('metadata_and_reviews-5.csv')
```
